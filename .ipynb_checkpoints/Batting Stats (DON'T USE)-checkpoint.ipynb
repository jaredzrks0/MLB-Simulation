```python
# Filter down to only regular season games

for year in all_pitches_dict:
    all_pitches_dict[year] = all_pitches_dict[year][all_pitches_dict[year].game_type == "R"]
    all_pitches_dict[year].game_date = all_pitches_dict[year].game_date.apply(lambda x: str(x).split(" ")[0])
```

```python
# Filter all pitches to only those resulting in a relavant play, and clean to only columns we might use later on
all_plays_dict = {}

for year in all_pitches_dict:
    # Filter all pitches to only those with an event
    all_plays_full_info = all_pitches_dict[year][pd.isna(all_pitches_dict[year].events) == False]
    
    # Filter all pitches with an event to only those types we care about
    all_relevant_plays = all_plays_full_info[all_plays_full_info.events.isin(relevant_play_types)]
    
    # Clean all the relevant plays and sort by date
    final_plays = all_relevant_plays[relevant_batting_columns].sort_values(by = "game_date").reset_index(drop = True)
    
    # Finally, add a new column that groups all the event types in to only the ones we will track
    final_plays["play_type"] = final_plays.events.apply(lambda x: play_type_dict[x])
    
    all_plays_dict[year] = final_plays
```

```python
def get_wind_direction(full_weather):
    """Given the full description, pull out the wind direction"""
    if full_weather != None:
        if "in" in "".join(full_weather.split("Wind")) or "In" in "".join(full_weather.split("Wind")):
            return "in" #full_weather.full_weather.split("mph ")[-1].split(' from')[0]
        elif "out" in "".join(full_weather.split("Wind")) or "Out" in "".join(full_weather.split("Wind")):
            return "out" #full_weather.full_weather.split("mph ")[-1].split(' to')[0]
        if "Left" in "".join(full_weather.split("Wind")) or "Right" in "".join(full_weather.split("Wind")):
            return full_weather.split("from ")[-1].strip(".")
    else:
        return "xyz"


```

## Create Batting Stats by Player

```python
all_plays_by_hand_combo = pkl.load(open("all_plays_by_hand_combo.pkl", "rb"))
```

```python
individual_batting_stats = {}

for year in all_plays_by_hand_combo.keys():
    individual_batting_stats[year] = {}
    year_dict = all_plays_by_hand_combo[year].copy()
    for pitbat_combo in year_dict.keys():
        
        pitbat_plays = year_dict[pitbat_combo].copy()
        
        pitbat_plays["type_counter"] = 1
        
        num_plays_by_play_type = pitbat_plays.groupby(by = ["game_pk", "batter", "play_type"]).count()
        num_plays_by_game = num_plays_by_play_type.groupby(by = ["game_pk", "batter"]).sum()
        
        num_plays_by_play_type = num_plays_by_play_type.reset_index()
        
        num_plays_by_play_type["play_share"] = num_plays_by_play_type.apply(lambda x: x.type_counter/num_plays_by_game.loc[x.game_pk, x.batter].type_counter, axis = 1)
        
        num_plays_by_play_type.game_date = num_plays_by_play_type.game_pk.apply(lambda x: pitbat_plays[pitbat_plays.game_pk == x].iloc[0].game_date)
        
        num_plays_by_play_type = num_plays_by_play_type[["game_pk", "batter", "play_type", "game_date", "type_counter", "play_share"]]
        
        individual_batting_stats[year][pitbat_combo] = num_plays_by_play_type
```
